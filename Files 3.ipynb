# Files:

- Reading a txt file
- writing a txt file

`f = open(filename, mode)`

Where the following mode is supported:

+ `r`: open an **existing** file for a *read* operation.
+ `w`: open an **existing** file for a *write* operation. If the file already contains some data then it will be **overridden**.
+ `a`: open an **existing** file for *append* operation. It won’t override existing data.
+ `r+`: To *read* and *write* data into the file. The previous data in the file will be **overridden**.
+ `w+`: To *write* and *read* data. It will **override** existing data.
+ `a+`: To *append* and *read* data from the file. It won’t override existing data.

# BBC news example:

```python
import os
```

```python
def filter_dirs(L, path):
    L1 = []
    for e in L:
        if os.path.isdir(os.path.join(path, e)):
            L1.append(e)
    return L1
```

```python
path = '../_data/_txt_data/bbc/'
dirs_list = filter_dirs(os.listdir(path), path)
print(dirs_list)
```

```python
from nltk.tokenize import word_tokenize
from datetime import datetime
```

```python
keywords_list = ['game', 'win', 'war', 'growth', 'oil', 'stock', 'dollar', 'euro', 'web', 'internet',
                 'love', 'travel', 'global']
```

```python
start=datetime.now()

data = []

for i in range(len(dirs_list)):

    folder_path = os.path.join(path, dirs_list[i])
#     print(dirs_list[i])
    files_list = sorted(os.listdir(folder_path))

    for f_name in files_list:
        try:
            file = open(os.path.join(folder_path, f_name), 'r')
            txt = file.read()
            
            word_list = word_tokenize(txt) # word_list = txt.split()
            feature_list = [dirs_list[i]]
            
            for k in keywords_list:
                c = 0
                for w in word_list:
                    if w.lower().find(k) != -1:
                        c += 1
                
                feature_list.append(c)
                
            data.append(feature_list)

        except:
            print(os.path.join(dirs_list[i], f_name) + ' is croped!')

print(datetime.now()-start)
```
